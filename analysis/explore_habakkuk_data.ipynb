```python
import jsonlib2 as json
from glob import glob
import gzip
from pandas import DataFrame, Series
```

```python
def get_data_from_store(dirname="/opt/habakkuk_data/", num_days=7):
    for fn in glob("%s/*"%dirname)[:num_days]:
        for line in gzip.open(fn):
            yield json.loads(line)
```

```python
data = [line for line in get_data_from_store()]
```
